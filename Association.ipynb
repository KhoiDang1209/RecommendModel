```python
import pandas as pd
import random

def create_transactions(main_categories, sub_categories, num_transactions=100):
    transactions = []

    for i in range(num_transactions):
        num_items = random.randint(1, 5)  # Random number of items per transaction
        transaction = []
        for _ in range(num_items):
            main_category = random.choice(main_categories)
            sub_category = random.choice(sub_categories[main_category])
            transaction.append(f"{main_category} - {sub_category}")
        
        transactions.append(transaction)

    # Convert to DataFrame
    transaction_df = pd.DataFrame({
        "TransactionID": range(1, num_transactions + 1),
        "Items": transactions
    })
    return transaction_df

def append_transactions(transaction_dfs):
    # Concatenate all transaction DataFrames
    overall_transaction_df = pd.concat(transaction_dfs, ignore_index=True)
    
    # Reassign TransactionID to ensure they are unique
    overall_transaction_df['TransactionID'] = range(1, len(overall_transaction_df) + 1)
    
    return overall_transaction_df

# Define main categories and subcategories
main_categories1 = ["accessories", "bags & luggage","beauty & health"]
sub_categories1 = {
    "accessories": ['Bags & Luggage', 'Fashion & Silver Jewellery', 'Handbags & Clutches', 'Jewellery', 'Watches', 'Sunglasses', 'Gold & Diamond Jewellery'],
    "bags & luggage": ['Travel Accessories', 'Rucksacks', 'Wallets', 'Backpacks', 'Suitcases & Trolley Bags', 'Travel Duffles'],
    "beauty & health": ['Make-up', 'Luxury Beauty', 'Value Bazaar']
}
main_categories2 = ["appliances", "home & kitchen"]
sub_categories2 = {
    "appliances": ['Heating & Cooling Appliances', 'Kitchen & Home Appliances', 'Washing Machines', 'Refrigerators'],
   "home & kitchen": ['Indoor Lighting', 'Home Improvement', 'Garden & Outdoors', 'Home Décor', 'Furniture', 'Home Storage', 'Kitchen & Dining', 'Sewing & Craft Supplies', 'Bedroom Linen', 'Home Furnishing', 'Kitchen Storage & Containers']
}
main_categories3 = ["beauty & health", "sports & fitness"]
sub_categories3 = {
    "beauty & health": ['Personal Care Appliances', 'Diet & Nutrition', 'Beauty & Grooming', 'Health & Personal Care', 'Household Supplies', 'Make-up', 'Luxury Beauty', 'Value Bazaar'],
    "sports & fitness": ['Fitness Accessories', 'Running', 'Cricket', 'Cycling', 'Badminton', 'Football', 'Strength Training', 'Yoga', 'Camping & Hiking', 'Cardio Equipment']
}
main_categories4 = ["grocery & gourmet foods", "home & kitchen"]
sub_categories4 = {
    "grocery & gourmet foods": ['Snack Foods', 'Coffee, Tea & Beverages'],
    "home & kitchen": ['Kitchen & Dining', 'Kitchen Storage & Containers']
}
main_categories5 = ["men's shoes", "men's clothing"]
sub_categories5 = {
    "men's clothing": ['Jeans', 'T-shirts & Polos', 'Shirts', 'Innerwear'],
    "men's shoes": ['Casual Shoes', 'Sports Shoes', 'Formal Shoes']
}
main_categories6 = ["women's clothing", "women's shoes"]
sub_categories6 = {
    "women's clothing": ['Ethnic Wear', 'Clothing', 'Lingerie & Nightwear', 'Western Wear'],
    "women's shoes": ['Fashion Sandals', 'Ballerinas', 'Shoes']
}
main_categories7 = ["kids' fashion", "toys & baby products"]
sub_categories7 = {
    "toys & baby products": ['Toys & Games', 'Baby Products', 'Diapers', 'Baby Bath, Skin & Grooming', 'Strollers & Prams', 'Nursing & Feeding', 'Toys Gifting Store', 'STEM Toys Store', 'International Toy Store'],
    "kids' fashion": ["Kids' Clothing", "Kids' Fashion", "Kids' Watches", 'School Bags', "Kids' Shoes", 'Baby Fashion']
}
main_categories8 = ["men's shoes", "women's shoes"]
sub_categories8 = {
    "men's shoes": ['Casual Shoes', 'Sports Shoes', 'Formal Shoes'],
    "women's shoes": ['Fashion Sandals', 'Ballerinas', 'Shoes']
}
main_categories9 = ["men's clothing", "women's clothing"]
sub_categories9 = {
    "men's clothing": ['Jeans', 'T-shirts & Polos', 'Shirts', 'Innerwear'],
    "women's clothing": ['Ethnic Wear', 'Clothing', 'Lingerie & Nightwear', 'Western Wear']
}
main_categories10 = ["home & kitchen", "tv, audio & cameras"]
sub_categories10 = {
    "home & kitchen": ['Indoor Lighting', 'Home Improvement', 'Garden & Outdoors', 'Home Décor', 'Furniture', 'Home Storage', 'Sewing & Craft Supplies', 'Bedroom Linen', 'Home Furnishing'],
    "tv, audio & cameras": ['Headphones', 'Cameras', 'Home Entertainment Systems', 'Speakers', 'Camera Accessories', 'Security Cameras', 'Televisions', 'Home Audio & Theater']
}
# Generate individual transaction DataFrames
df1 = create_transactions(main_categories1, sub_categories1, num_transactions=500)
df2 = create_transactions(main_categories2, sub_categories2, num_transactions=500)
df3 = create_transactions(main_categories3, sub_categories3, num_transactions=500)
df4 = create_transactions(main_categories4, sub_categories4, num_transactions=250)
df5 = create_transactions(main_categories5, sub_categories5, num_transactions=250)
df6 = create_transactions(main_categories6, sub_categories6, num_transactions=250)
df7 = create_transactions(main_categories7, sub_categories7, num_transactions=500)
df8 = create_transactions(main_categories8, sub_categories8, num_transactions=250)
df9 = create_transactions(main_categories9, sub_categories9, num_transactions=250)
df10 = create_transactions(main_categories10, sub_categories10, num_transactions=500)

# Combine all transaction DataFrames into an overall DataFrame
overall_transaction_df = append_transactions([df1,df2,df3,df4,df5,df6,df7,df8,df9,df10])
# Display the combined DataFrame
print(overall_transaction_df)
```

```python
from mlxtend.preprocessing.transactionencoder import TransactionEncoder

from mlxtend.frequent_patterns import association_rules, fpgrowth
transactions_list = overall_transaction_df['Items'].tolist()
te = TransactionEncoder()
te_ary = te.fit(transactions_list).transform(transactions_list)
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)
# Step 2: Use Apriori to find frequent itemsets
frequent_itemsets = fpgrowth(df_encoded, min_support=0.005, use_colnames=True)
frequent_itemsets_size = len(frequent_itemsets)  # Calculate the number of itemsets
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.5)

rules['antecedents'] = rules['antecedents'].apply(lambda x: ', '.join(list(x)))
rules['consequents'] = rules['consequents'].apply(lambda x: ', '.join(list(x)))

# Step 4: Display the rules
rules.to_csv("association_rules.csv", index=False)
print(rules)
```

```python
import pandas as pd

# Load data
rule = pd.read_csv("association_rules.csv")
product = pd.read_csv("test_data.csv")

def get_top_associated_category(id, data, rules, n=5):
    # Retrieve item information
    item_info = data[data['id'] == id]
    if item_info.empty:
        print("Item not found in dataset.")
        return []

    # Construct item category combination
    item_category = f"{item_info.iloc[0]['main_category']} - {item_info.iloc[0]['sub_category']}"
    print(f"Looking for item category: {item_category}")

    # Filter rules to find relevant antecedents
    relevant_rules = rules[rules['antecedents'].apply(lambda x: item_category in x)]
    print(f"Relevant rules found: {len(relevant_rules)}")

    # If no rules, return empty list
    if relevant_rules.empty:
        print("No association rules found for this item.")
        return []

    # Sort rules and get top associated categories
    relevant_rules = relevant_rules.sort_values(by='lift', ascending=False).head(n)
    top_associated_categories = set()

    for _, row in relevant_rules.iterrows():
        for category_set in [row['antecedents'], row['consequents']]:
            if isinstance(category_set, str):  # Ensure it's a string
                for category in category_set.split(", "):  # Split by delimiter
                    if category != item_category:  # Exclude the original category
                        top_associated_categories.add(category)

    # Convert to list and return top N categories
    top_categories_list = list(top_associated_categories)[:n]
    print(f"Top associated categories: {top_categories_list}")
    return top_categories_list

def get_top_rated_items(data, associated_categories, top_n=4):
    top_items = []
    # Iterate through each category and get top-rated items
    for category in associated_categories:
        # Split the main and subcategories
        try:
            main_category, sub_category = category.split(" - ")
        except ValueError:
            print(f"Invalid category format: {category}")
            continue
        # Filter items within the given category
        category_items = data[(data['main_category'] == main_category) &
                              (data['sub_category'] == sub_category)]
        # Get top items based on ratings
        top_rated_items = category_items.sort_values(by='ratings', ascending=False).head(top_n)
        # Add to the top items list
        top_items.extend(top_rated_items['id'].tolist())
    return top_items

# Example usage
top_associated_categories = get_top_associated_category('wcl099915', product, rule)
print("Top associated category:", top_associated_categories)
top_rated_items = get_top_rated_items(product, top_associated_categories)
print("Top rated items from associated categories:", top_rated_items)
```
